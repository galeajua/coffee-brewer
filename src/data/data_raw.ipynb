```python
# next up, we want to standardize the values for the 'Altitude' column
# altitude values should be represented by a single numeric value representing meters, for example 1350

import numpy as np
import re


def clean_altitude(value):

    if pd.isna(value):
        return np.nan
    
    value = str(value).lower().strip()
    
    # Case 1: Match ranges with optional spaces and hyphen (e.g., "1000 - 1500", "1000-1500 m")
    match_range = re.match(r'(\d+)\s*-\s*(\d+)', value)
    if match_range:
        min_alt, max_alt = map(int, match_range.groups())
        return (min_alt + max_alt) / 2  # Calculate the mean of the range
    
    # Case 2: Match "Meters above sea level" pattern (e.g., "meters above sea level: 1.872")
    match_meters = re.search(r'meters above sea level: ([\d.]+)', value)
    if match_meters:
        # Convert the matched altitude to a float
        return float(match_meters.group(1))
    
    # Case 3: Match approximate or single number (e.g., "~350" or "≈350")
    match_approx = re.match(r'[~≈]?(\d+)', value)
    if match_approx:
        return int(match_approx.group(1))
    
    # Case 4: Match single number within the text (e.g., "1200m")
    match_single = re.search(r'(\d+)', value)
    if match_single:
        return int(match_single.group(1))
    
    # Case 5: Unrecognized formats (e.g., "TEST")
    return np.nan


df['Altitude_clean'] = df['Altitude'].apply(clean_altitude)
```

```python
# Identify rows where 'Altitude' is non-null but 'Altitude_clean' is NaN
unmatched_values = df[df['Altitude'].notna() & df['Altitude_clean'].isna()]

# Display these rows to understand the issue
print(unmatched_values[['Altitude', 'Altitude_clean']])
```

```python
# after few iterations and adjusting the 'clean_altitude'-method, it seems that the remaining non-null values for 'Altitude' were transformed correctly to NaN
# now we can replace the original 'Altitude' column with the new'Altitude_clean' and downsize it to float32

df['Altitude_clean'] = df['Altitude_clean'].apply(lambda x: x * 1000 if x < 10 else x)

df['Altitude'] = df['Altitude_clean']
df = df.drop(columns=['Altitude_clean'])

df['Altitude'] = df['Altitude'].astype('float32')
```
